```python
soup = BeautifulSoup(page.text, "html.parser")
```

```python
soup
```

```python
proftags = soup.findAll("li", {"class": "toclevel-2 tocsection-2" })
proftags
```

```python
[]
for mytag in proftags:
    print(mytag.get_text())
```

```python
list_=[]

for X in range(22):
    for Y in range(1,3):
    #print(X)
        Y = "toclevel-"+str(Y)+ " tocsection-"+str(X)
        list_.append(Y)
    #print(Y)
    
    
```

```python
for i in list_:
    proftags = soup.findAll("li", {"class": i })
    for mytag in proftags:
        print(mytag.get_text())    
```
